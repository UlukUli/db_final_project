# AI Agent for Hotel Booking Database

## Project Overview
This notebook implements an AI-powered assistant for working with a hotel bookings relational database.
The agent should:
- Understand natural language questions.
- Generate SQL queries to the database.
- Return analytical results and insights.

## Business Problems

1. Booking Cancellations

Goal: Identify which types of customers are more likely to cancel their bookings and what factors contribute to higher cancellation rates.

2. Hotel Occupancy by Month

Goal: Determine which months have higher booking volumes and compare occupancy trends between the City Hotel and the Resort Hotel.

3. Popular Room Types

Goal: Analyze which room types are booked most frequently and how this varies by customer type and season.

4. Revenue by Market Segment

Goal: Evaluate which market segments (e.g., Corporate, Direct, Online) generate the highest revenue for the hotel.

5. Impact of Meal Plans on Bookings

Goal: Understand which meal plans (BB, HB, FB, SC) are in highest demand and how meal preferences relate to customer types and booking behavior.

- Architect – database structure and ERD
- Analyst – business problems, metrics, EDA
- SQL Developer – tables, indexes, views
- AI Engineer – AI agent, text-to-SQL, analytics in Python

```python
SCHEMA_DESCRIPTION = """
TABLE: hotel_bookings (one row per hotel booking)

Columns:
- hotel: type of hotel (Resort Hotel, City Hotel)
- is_canceled: booking canceled (1) or not (0)
- lead_time: days between booking and arrival
- arrival_date_year: year of arrival
- arrival_date_month: month of arrival (text)
- arrival_date_week_number: ISO week number
- arrival_date_day_of_month: day of month (1–31)
- stays_in_weekend_nights: number of weekend nights stayed
- stays_in_week_nights: number of week nights stayed
- adults: number of adults
- children: number of children
- babies: number of babies
- meal: meal type (BB, HB, FB, SC)
- country: country of origin
- market_segment: booking channel (Online TA, Offline TA, Corporate, Direct...)
- distribution_channel: distribution type (TA/TO, Direct, Corporate, GDS...)
- is_repeated_guest: 1 if guest stayed before, else 0
- previous_cancellations: number of past cancellations by this customer
- reserved_room_type: reserved room code (A,B,C...)
- assigned_room_type: assigned room code
- booking_changes: number of changes to booking
- deposit_type: No Deposit / Non Refund / Refundable
- agent: agent id
- company: company id
- days_in_waiting_list: how long booking waited
- customer_type: Transient, Contract, Group...
- adr: Average Daily Rate (price per night per occupied room)
- required_car_parking_spaces: parking spaces
- total_of_special_requests: count of special requests
"""
```

```python
reference_queries = {
    "cancellation_rate_by_hotel": """
        SELECT
            hotel,
            AVG(is_canceled) AS cancellation_rate
        FROM hotel_bookings
        GROUP BY hotel;
    """,

    "bookings_by_month": """
        SELECT
            arrival_date_month,
            COUNT(*) AS total_bookings
        FROM hotel_bookings
        GROUP BY arrival_date_month
        ORDER BY total_bookings DESC;
    """,

    "popular_room_types": """
        SELECT
            reserved_room_type,
            COUNT(*) AS total_bookings
        FROM hotel_bookings
        GROUP BY reserved_room_type
        ORDER BY total_bookings DESC;
    """,

    "revenue_by_market_segment": """
        SELECT
            market_segment,
            SUM(adr * (stays_in_weekend_nights + stays_in_week_nights)) AS total_revenue
        FROM hotel_bookings
        WHERE is_canceled = 0
        GROUP BY market_segment
        ORDER BY total_revenue DESC;
    """,

    "adr_by_month_and_hotel": """
        SELECT
            hotel,
            arrival_date_month,
            AVG(adr) AS avg_adr
        FROM hotel_bookings
        WHERE is_canceled = 0
        GROUP BY hotel, arrival_date_month
        ORDER BY hotel, arrival_date_month;
    """
}
```

```python
def build_sql_prompt(user_question: str) -> str:
    return f"""
You are an expert MySQL analyst for a hotel bookings database.

Database schema:
{SCHEMA_DESCRIPTION}

Task:
- Given the user question, write ONE valid MySQL SELECT query.
- Use only existing columns.
- Do NOT explain the query, output ONLY raw SQL.

User question: "{user_question}"

SQL:
""".strip()
```

```python
def fake_generate_sql(user_question: str) -> str:
    q = user_question.lower()

    if "cancel" in q:
        return reference_queries["cancellation_rate_by_hotel"]
    if "month" in q:
        return reference_queries["bookings_by_month"]
    if "room" in q:
        return reference_queries["popular_room_types"]
    if "segment" in q or "market" in q:
        return reference_queries["revenue_by_market_segment"]
    if "adr" in q or "price" in q:
        return reference_queries["adr_by_month_and_hotel"]

    # fallback — временно
    return "SELECT 1;"
```

```python
question = "What is the cancellation rate by hotel type?"
print(fake_generate_sql(question))
```
